```python
__author__ = "Xyba"
__copyright__ = "2017 - Project Xyba"

try:
    from Tkinter import *
    import ttk
except ImportError:
    from tkinter import *
    import tkinter.ttk as ttk
import random as rd
import warnings
warnings.filterwarnings("ignore")

class Randomizer:
    def __init__(self, master, title, icon, version):
        self.master = master
        self.title = title + ' v%s'%(version)
        self.icon = icon
        
        self.master.title(self.title)
        self.master.wm_iconbitmap(icon)
        self.master.resizable(width = False, height = False)
        
        self.components()
        
        self.menus(self.master)
        
        master.bind('<Return>', self.random)
        
    def components(self):
        self.frame = ttk.Frame(self.master)
        self.frame.grid(row = 0, column = 0)
        
        self.vcmdA = self.master.register(self.validateA)
        self.vcmdB = self.master.register(self.validateB)
        
        self.lowerbound = Label(self.master, text = 'Lower Bound', font = (('Cambria', '12')))
        self.lowerbound.grid(row = 1, column = 0, sticky = W)
        
        self.upperbound = Label(self.master, text = 'Upper Bound', font = (('Cambria', '12')))
        self.upperbound.grid(row = 2, column = 0, sticky = W)
        
        self.entryA = Entry(self.master, validate = 'key', validatecommand = (self.vcmdA, '%P'),
                            font = (('Cambria', '12')), justify = 'center')
        self.entryB = Entry(self.master, validate = 'key', validatecommand = (self.vcmdB, '%P'),
                            font = (('Cambria', '12')), justify = 'center')
        
        self.entryA.grid(row = 1, column = 1, sticky = W + E)
        self.entryB.grid(row = 2, column = 1, sticky = W + E)
        
        self.random_button = Button(self.master, text = 'Randomize', font = (('Cambria', '16')), command = self.random)
        self.random_button.grid(row = 3, column = 1, sticky = W + E)
        
        self.result = Label(self.master, text = 'Result\t:', font = (('Cambria', '16')))
        self.result.grid(row = 4, column = 0, sticky = W)
        
        self.cmessage = 'Created by Xyba'
        self.creator_text = StringVar()
        self.creator_text.set(self.cmessage)
        self.creator = Label(self.master,
                             textvariable = self.creator_text,
                             font = (('Cambria', '16')),
                             relief = SUNKEN)
 
        self.creator.grid(row = 5, column = 0, columnspan = 3, sticky = W + E)
    
    def menus(self, master):
        menu = Menu(master)
        master.config(menu = menu)
        
        fileMenu = Menu(menu, tearoff = 0)
        menu.add_cascade(label = 'File', menu = fileMenu)
        fileMenu.add_command(label = 'Exit', command = master.destroy)
    
        aboutMenu = Menu(menu, tearoff = 0)
        menu.add_cascade(label = 'About', menu = aboutMenu)
        aboutMenu.add_command(label = 'About', command = self.About)

    def validateA(self, userinput):
        if not userinput:
            self.entryA = None
            return True

        try:
            entryA = int(userinput)
            if 1 <= entryA <= 2**31 - 1:
                self.entryA = entryA
                return True
            else:
                return False

        except ValueError:
            return False
    
    def validateB(self, userinput):
        if not userinput:
            self.entryB = None
            return True

        try:
            entryB = int(userinput)
            if 1 <= entryB <= 2**31 - 1:
                self.entryB = entryB
                return True
            else:
                return False

        except ValueError:
            return False
    
    def random(self, event = None):
        try:
            if self.entryA != None and self.entryB != None:
                if self.entryA < self.entryB:
                    self.outmessage = str(rd.randint(self.entryA, self.entryB))
                    self.outmessage_text = StringVar()
                    self.outmessage_text.set(self.outmessage)
                    self.output = Label(self.master, textvariable = self.outmessage_text, font = (('Cambria', '16')))

                    self.output.grid(row = 4, column = 1, sticky = W + E)

                    return True

            return False
        
        except TypeError:
            return False

    def About(self):
        toplevel = Toplevel()
        toplevel.title(Title + ' v%s: About'%(ver))
        toplevel.iconbitmap(icon)
        toplevel.geometry('500x200')
        toplevel.resizable(width = False, height = False)
        label = Label(toplevel,
                      text = 'This program is made by Xyba, student of Mathematics UI 2016\n\n\n\n\n2017 - Xyba Project',
                      font = (('Cambria', '12')),
                      height = 100,
                      width = 100)
        label.pack()

Title = 'Randomizer'
icon = 'Xyba_Logo.ico' # 'MiniNoire.ico'
ver = '0.9'

def main():
    root = Tk()
    Randomizer(root, Title, icon, ver)
    root.mainloop()

if __name__ == '__main__':
    main()
```
